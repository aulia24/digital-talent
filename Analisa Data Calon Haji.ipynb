```python
data_calonhaji= pd.read_csv('jumlahcalonjemaahhajimenurutprofesidankotaadministrasi2011.csv', delimiter = ',' ,encoding = "ISO-8859-1")
```

# 1. Deskripsi Data

Data di atas merupakan representasi dari calon jemaah haji di DKI Jakarta pada tahun 2011 berdasarkan profesi yang disajikan dalam bentuk tabel. Profesinya terbagi menjadi 9, yaitu PNS, TNI/Polri, Pedagang,Petani/Nelayan, Pegawai Swasta, Pelajar/Mahasiswa, Ibu Rumah Tangga, Pegawai BUMN, dan Lainnya. Data ini terbagi merata di 5 kota di DKI Jakarta yaitu Jakarta Selatan, Jalarta Utara, jakarta Barat, Jakarta Timur, dan Jakarta Pusat. 

# 2. Statistik

```python
import statistics

s = data_calonhaji.jumlah.sum()
mi = data_calonhaji.jumlah.min()
ma = data_calonhaji.jumlah.max()
me = data_calonhaji.jumlah.mean()

print("Jumlah Calon Jemaah Haji di DKI Jakarta 2011 : ", s)
print("Jumlah Terkecil Calon Jemaah Haji di DKI Jakarta 2011 : ",mi)
print("Jumlah Terbesar Calon Jemaah Haji di DKI Jakarta 2011 : ",ma)
print("Jumlah Rata-rata Jemaah Haji di DKI Jakarta 2011 : ",me)
```

# 3. Visualisasi Data

```python
# Visualisasi 1

ax = data_calonhaji.groupby('profesi').mean()['jumlah']
ax = ax.plot(kind='bar', title ="Profesi Calon Jemaah Haji di DKI Jakarta Tahun 2011", figsize=(20,20), legend=True, fontsize=12)
ax.set_xlabel("Profesi", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

```python
# Visualisasi 2

ax = data_calonhaji.groupby('nama_kabkota').mean()['jumlah']
ax = ax.plot(kind='bar', title ="Jumlah Calon Jemaah Haji di DKI Jakarta Tahun 2011", figsize=(20,20), legend=True, fontsize=12)
ax.set_xlabel("Nama Kota", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```
